```
# Create the Spider class
class BBC_Spider(scrapy.Spider):
  name = "bbcyoruba_news_spider"
  # start_requests method
  def start_requests(self):
    yield scrapy.Request(url = 'https://www.bbc.com/yoruba',
                         callback = self.parse_front)
  # First parsing method
  def parse_front(self, response):
    news_blocks = response.xpath('//div[contains(@class, "css-8tq3w8 ezetrkd0")]')  
    news_links = news_blocks.xpath('.//a/@href')
    links_to_follow = news_links.extract()
    for url in links_to_follow:
      yield response.follow(url = 'https://www.bbc.com' + url,
                            callback = self.parse_pages)
  # Second parsing method
  def parse_pages(self, response):
    news_title = response.xpath('//title/text()')
    news_title_ext = news_title.extract_first().strip()
    news_article = response.xpath('//p[contains(@class, "css-qg6qrj-Paragraph e1cc2ql70")]/text()')
    news_article_ext = [t.strip() for t in news_article.extract()]
    dc_dict[ news_title_ext ] = news_article_ext

# Initialize the dictionary **outside** of the Spider class
dc_dict = dict()

# Run the Spider
process = CrawlerProcess()
process.crawl(BBC_Spider)
process.start()

# Print a preview of courses
print(dc_dict)
```
